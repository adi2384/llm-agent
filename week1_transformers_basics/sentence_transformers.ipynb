```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load a small, fast model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example paragraph from a news-style article (split into sentences)
paragraph = """Container shipping has revolutionized global trade.
It began in 1956 when Malcolm McLean loaded 58 containers onto a converted oil tanker.
This innovation drastically reduced loading times at ports.
Before containers, goods were handled piece by piece, causing long delays.
Standard containers could be moved easily between ships, trucks, and trains.
Over time, ports were redesigned to handle these metal boxes more efficiently.
This led to lower transportation costs and more reliable delivery times.
Today, container shipping carries over 90% of global goods by volume.
It remains one of the most important pillars of the modern global economy."""


# Turn it into a list of sentences
sentences = paragraph.strip().split('\n')

# Create embeddings
embeddings = model.encode(sentences)

# Compute cosine similarity matrix
cos_sim = cosine_similarity(embeddings)

# Show top 5 similar sentences for each one
for i, sent in enumerate(sentences):
    print(f"\n🔍 For: '{sent}'")
    sim_scores = list(enumerate(cos_sim[i]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    top_k = [sentences[idx] for idx, score in sim_scores[1:6]]  # Skip itself
    for match in top_k:
        print(f"  ➤ {match}")
```
